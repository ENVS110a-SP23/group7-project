```python
pollution = pd.read_csv("Data/pollution_us_2000_2016.csv")
```

```python
pollution.columns
pollution.shape
pollution.head()
```

```python
len(pd.unique(pollution['State']))
```

```python
pollution_means = pollution[['Date Local','NO2 Mean','O3 Mean', 'SO2 Mean', 'CO Mean']]

#pollution_lite = pollution_means.groupby(['Date Local']).mean().reset_index()
pollution_means
```

```python
# transportation.reset_index(inplace=True)

# transportation.drop(columns=["East Coast Production of Gasoline (Thousand Barrels per Day)","East Coast Production of Jet Fuel (Thousand Barrels per Day)"],inplace=True)
```

```python
pollution_means['Date Local'] = pd.to_datetime(pollution_means['Date Local']) + pd.to_timedelta(7, unit='d')
```

```python
pollution_lite = pollution_means.groupby([pd.Grouper(key='Date Local', freq='W-FRI')]).mean().reset_index().sort_values('Date Local')
pollution_lite
```

```python
for column in pollution_lite.columns:
    
    pollution_lite.rename(columns={column:column.split(" ")[0]},inplace=True)

pollution_lite["Date"] = pd.to_datetime(pollution_lite["Date"])

pollution_lite
```

```python
pollution_and_fuels = transportation.merge(pollution_lite,how='left',on='Date').dropna()

pollution_and_fuels.drop(columns=["East Coast Production of Gasoline (Thousand Barrels per Day)","East Coast Production of Jet Fuel (Thousand Barrels per Day)"])
```

```python
pollution_and_fuels.drop(["Date","East Coast Production of Gasoline (Thousand Barrels per Day)","East Coast Production of Jet Fuel (Thousand Barrels per Day)","U.S. Gasoline Demand (Thousand Barrels per Day)"],axis=1,inplace=True)
pollution_and_fuels.rename(columns={"U.S. Production of Gasoline (Thousand Barrels per Day)":"Gasoline Production","U.S. Production of Jet Fuel (Thousand Barrels per Day)":"Jet fuel production"},inplace=True)

```

```python
pollution_and_fuels.dropna().corr()
```

```python
sns.heatmap(pollution_and_fuels.corr(),annot=True)
```
